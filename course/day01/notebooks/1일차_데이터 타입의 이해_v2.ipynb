# 📋 데이터 타입의 이해

데이터 분석 과정에서 가장 기본이 되는 것은 다루고 있는 데이터의 타입과 성격을 정확히 이해하는 것입니다.   
환자의 나이, 성별, 혈압, 진료과 등 각기 다른 특성을 가진 데이터들은 서로 다른 분석 방법이 필요합니다.

데이터 타입에 따라 적절한 분석 방법과 시각화 기법이 달라지므로, 올바른 데이터 타입 분류는 성공적인 분석의 핵심입니다.

## 왜 데이터 타입 구분이 중요한가?

데이터 타입을 올바르게 구분하지 않으면 잘못된 분석 결과를 얻을 수 있습니다.   
각 타입별로 적용 가능한 분석 방법이 다르기 때문입니다.



### 잘못된 분석 예시

**범주형 변수를 숫자형으로 잘못 해석하는 경우**



```python
# 병실 번호 (명목형 - 숫자로 표현되지만 범주형)
room_numbers = [101, 205, 312, 108, 220, 101, 205, 101, 312, 207]
```

**❌ 잘못된 분석**


```python
# 평균 병실 번호를 계산하면?
average_room = sum(room_numbers) / len(room_numbers)
print(f"평균 병실: {average_room}호")
# → 187.2호라는 병실은 존재하지 않음! 의미 없는 계산
```

**✅ 올바른 분석**


```python
# ✅ 올바른 접근: 명목형으로 취급
import pandas as pd  # 데이터 분석 라이브러리

room_series = pd.Series(room_numbers)  # 리스트를 pandas Series로 변환
room_counts = room_series.value_counts()  # 각 값의 빈도수 계산 (내림차순 정렬)
most_used_room = room_series.mode()[0]  # 최빈값(가장 자주 나타나는 값) 추출

print("병실 사용 현황:")
print(room_counts)
print(f"가장 자주 사용되는 병실: {most_used_room}호")
# → 실제 존재하는 병실 번호로 의미 있는 분석
```

**환자 만족도 분석**



```python
# 만족도 점수 데이터 (순서형 - 숫자로 표현되지만 범주형)
satisfaction_scores = [
    1,
    2,
    3,
    4,
    5,
    3,
    4,
    4,
    5,
    3,
]  # 1:매우불만, 2:불만, 3:보통, 4:만족, 5: 매우만족

# 단순히 만족도 수준을 나타내는 라벨이지, 실제 수치적 의미를 갖는 숫자가 아님
```

```python
# ❌ 잘못된 접근: 연속형으로 취급
average_satisfaction = sum(satisfaction_scores) / len(satisfaction_scores)
print(f"평균 만족도: {average_satisfaction:.1f}점")
# → 3.4점이라는 선택지는 설문에 존재하지 않음!
```

```python
# ✅ 올바른 접근: 순서형으로 취급
satisfaction_series = pd.Series(satisfaction_scores)  # 리스트를 pandas Series로 변환
median_satisfaction = satisfaction_series.median()  # 중간값 계산 (순서형에서 중심 경향 측정)
mode_satisfaction = satisfaction_series.mode()[0]  # 최빈값 추출 (가장 자주 선택된 만족도)
satisfaction_dist = satisfaction_series.value_counts().sort_index()  # 빈도수를 점수 순서대로 정렬

print("만족도 분포:")
print(satisfaction_dist)
print(f"중간 만족도: {median_satisfaction}점")
print(f"가장 많은 응답: {mode_satisfaction}점")
# → 실제 선택 가능한 점수로 의미 있는 해석
```

## 데이터 타입에 따른 분석 방법

데이터 타입을 정확히 파악하면 적절한 분석 방법을 선택할 수 있습니다.



| 구분 | 숫자형 데이터 (Numerical) | 범주형 데이터 (Categorical) |
|------|---------------------------|---------------------------|
| **정의** | 숫자로 표현되며 수학적 계산이 가능한 데이터 | 그룹이나 범주로 분류되는 질적 데이터 |
| **의료 예시** | 나이(45세), 체중(70kg), 혈압(120mmHg), BMI(22.5) | 성별(남/여), 혈액형(A/B/O/AB), 진료과(내과/외과) |
| **특징** | • 수치적 의미<br>• 대소 비교 가능<br>• 연산 결과가 의미 있음 | • 분류/구분 목적<br>• 라벨링 역할<br>• 연산 결과가 무의미 |
| **가능한 분석** | 평균, 중간값, 표준편차, 상관관계 | 빈도, 비율, 최빈값, 교차분석 |
| **시각화** | 히스토그램, 산점도, 박스플롯 | 막대그래프, 파이차트 |
| **판단 기준** | "평균을 구하는 것이 의미가 있는가?" → **Yes** | "평균을 구하는 것이 의미가 있는가?" → **No** |
| **세부 분류** | • 연속형: 키 175.5cm<br>• 이산형: 환자 5명 | • 명목형: 혈액형 A형<br>• 순서형: 만족도 상/중/하 |





### 타입별 특징

- **연속형 데이터**: 모든 값이 수학적으로 의미가 있고, 값들 사이의 거리가 일정하므로 평균과 표준편차를 통해 분포의 중심과 퍼짐을 정확히 측정할 수 있습니다.

- **이산형 데이터**: 정수값만 존재하고 개수를 세는 특성상, 최빈값(가장 자주 나타나는 값)과 범위가 데이터의 특성을 더 잘 나타냅니다.

- **명목형 데이터**: 범주 간에 순서가 없어 평균이나 중간값이 의미가 없으므로, 각 범주의 빈도와 비율을 통해서만 분포를 파악할 수 있습니다.

- **순서형 데이터**: 순서는 있지만 범주 간 간격이 일정하지 않아 평균보다는 중간값이 더 안정적이고 의미 있는 중심 경향을 나타냅니다.



### 타입별시각화 방법

| 데이터 타입 | 적합한 그래프 | 부적합한 그래프 |
|------------|--------------|----------------|
| **연속형** | 히스토그램, 박스플롯, 산점도 | 막대그래프 |
| **이산형** | 막대그래프, 점 그래프 | 히스토그램 |
| **명목형** | 막대그래프, 파이차트 | 히스토그램, 선 그래프 |
| **순서형** | 막대그래프(순서대로), 누적그래프 | 파이차트 |
