# 📋 데이터 타입의 이해

데이터 분석을 시작할 때 가장 먼저 확인할 것은 “이 컬럼이 어떤 종류의 데이터인가?”입니다.  
환자의 나이, 성별, 혈압, 진료과처럼 서로 다른 성격의 데이터는 분석 방법도 달라집니다.

의료 데이터에서는 검사 결과, 설문 응답, 진료 기록, 환자 기본 정보가 함께 섞여 있으므로 데이터 타입을 먼저 구분하는 습관이 중요합니다.

- `45`는 숫자입니다. 더하기, 평균 계산, 대소 비교가 가능합니다.
- `"45"`는 문자열입니다. 화면에 표시되는 글자라서 바로 평균을 낼 수 없습니다.

## 숫자형 데이터 유형

숫자형 데이터는 값의 특성에 따라 **연속형**과 **이산형**으로 나뉩니다.



### 연속형 데이터 (Continuous Data)

이론적으로 **무한히 세분화** 가능하여 데이터와 데이터 간 **끊어짐이 없는 연속적인 값**을 가지는 데이터입니다.   
소수점 값이 의미를 가지며, 임의의 두 값 사이에 항상 다른 값이 존재할 수 있습니다.



```python
# 키 데이터 (cm)
heights = [175.2, 162.7, 180.1, 158.5, 170.8]

# 체중 데이터 (kg)
weights = [70.5, 58.3, 85.7, 52.1, 75.9]

# 체온 데이터 (℃)
temperatures = [36.5, 37.2, 36.8, 37.1, 36.7]

# BMI 데이터
bmi_values = [22.9, 21.5, 26.2, 20.8, 25.1]

# 혈압 데이터 (mmHg)
blood_pressure = [120.5, 140.2, 110.8, 115.3, 135.7]
```

#### 시간 순서가 붙으면 시계열 데이터가 됩니다

연속형 데이터는 한 번만 측정할 수도 있지만, 같은 대상을 시간 순서대로 반복 측정할 수도 있습니다.  
예를 들어 하루별 혈압, 시간대별 혈당, 주차별 체중 변화는 모두 **시간 순서가 붙은 연속형 데이터**입니다.

이런 데이터를 보통 **시계열 데이터(Time Series)**라고 부릅니다.  
시계열은 “연속형/이산형/범주형”처럼 값 자체의 타입을 나누는 말이라기보다, **시간 축을 따라 관찰된 데이터 구조**에 가깝습니다.

- 측정값: 혈압, 혈당, 체온처럼 연속형일 수 있음
- 시간 정보: 날짜, 시간, 주차, 월처럼 순서를 나타냄
- 대표 그래프: 시간 흐름을 보여주는 선 그래프


```python
# 시계열 형태의 연속형 데이터 예시
measurement_dates = ["2026-05-01", "2026-05-02", "2026-05-03", "2026-05-04"]
systolic_bp_by_day = [140.0, 138.5, 135.2, 132.8]

for date, bp in zip(measurement_dates, systolic_bp_by_day):
    print(f"{date}: 수축기 혈압 {bp} mmHg")
```

**특징**
- 소수점 값 존재
- 구간별 분포 분석
- 정규분포 등 연속 확률분포 적용
- 회귀분석에 적합

** 회귀분석: 여러 변수들이 특정 변수에 미치는 영향력의 크기와 방향을 수치적으로 분석하여, 변수 간의 관계를 설명하고 미래 값을 예측하는 통계 기법

## 범주형 데이터 유형

범주형 데이터는 범주 간의 **순서 관계**에 따라 **명목형**과 **순서형**으로 구분됩니다.



### 범주형과 명목형은 같은 말인가?

완전히 같은 말은 아닙니다. **범주형 데이터**가 더 큰 묶음이고, **명목형 데이터**는 범주형 데이터의 한 종류입니다.

| 구분 | 의미 | 예시 |
|---|---|---|
| 범주형 데이터 | 몇 개의 그룹이나 라벨로 나뉘는 데이터 전체 | 성별, 혈액형, 진료과, 만족도, 통증 정도 |
| 명목형 데이터 | 범주형 중에서 순서가 없는 데이터 | 성별, 혈액형, 진료과 |
| 순서형 데이터 | 범주형 중에서 순서가 있는 데이터 | 만족도, 통증 정도, 병기 |

즉, **명목형은 범주형이다**라고 말할 수 있지만, **범주형이 모두 명목형은 아닙니다.**  
범주형 안에는 명목형과 순서형이 함께 있습니다.


### 순서형 데이터 (Ordinal Data)

**특징**

- 중간값, 백분위수 계산 가능
- 순위 기반 분석 방법 적용
- 스피어만 상관계수 등 사용


### 명목형 데이터 (Nominal Data)

범주 간에 **순서나 서열이 없는** 데이터입니다. 단순히 그룹을 구분하는 역할만 합니다.

#### 카이제곱 검정은 언제 쓰나?

명목형 데이터끼리는 평균을 비교할 수 없습니다. 대신 **각 범주의 빈도표**를 보고 두 범주형 변수가 서로 관련이 있는지 살펴봅니다.

대표 예시:

> 진료과(내과/외과/소아과)와 재방문 여부(예/아니오)는 서로 관련이 있을까?

이럴 때 먼저 교차표를 만들고, 더 나아가면 **카이제곱 검정(Chi-square test)**으로 두 범주형 변수의 독립성을 확인할 수 있습니다.


```python
import pandas as pd

# 카이제곱 검정으로 이어질 수 있는 대표 예시: 교차표 만들기
visit_data = pd.DataFrame(
    {
        "진료과": ["내과", "내과", "내과", "외과", "외과", "소아과", "소아과", "소아과"],
        "재방문여부": ["예", "예", "아니오", "예", "아니오", "아니오", "아니오", "예"],
    }
)

contingency_table = pd.crosstab(visit_data["진료과"], visit_data["재방문여부"])
contingency_table
```

위 표처럼 **관찰된 빈도**를 정리한 뒤, 각 칸의 차이가 우연으로 보기 어려운지 판단할 때 카이제곱 검정을 사용합니다.

이번 노트북에서는 “언제 쓰는지”만 이해하면 충분합니다. 실제 검정 계산은 통계 분석 단계에서 다룹니다.

참고 링크:
- [SciPy `chi2_contingency` 공식 문서](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.chi2_contingency.html): 교차표 기반 카이제곱 독립성 검정 함수
- [NIST Chi-Square Distribution 설명과 그래프](https://www.itl.nist.gov/div898/handbook/eda/section3/eda3666.htm): 카이제곱 분포 그림이 포함된 참고 자료


## 왜 데이터 타입 구분이 중요한가?

데이터 타입을 잘못 이해하면 코드가 에러를 내거나, 더 위험하게는 **그럴듯하지만 의미 없는 분석 결과**를 만들 수 있습니다.

가장 쉬운 예는 숫자와 문자열입니다.  
`45`는 평균을 낼 수 있지만, `"45세"`는 그대로 평균을 낼 수 없습니다.  
`101`이라는 병실 번호도 숫자처럼 보이지만, 평균 병실 번호를 구하는 것은 의미가 없습니다.

따라서 분석 전에 먼저 질문해야 합니다.

> 이 값은 계산할 숫자인가, 아니면 구분하기 위한 이름표인가?


```python
# 숫자처럼 보여도 데이터 의미에 따라 다르게 다뤄야 함
age = 45  # 나이: 계산 가능한 숫자형
room_number = "101"  # 병실 번호: 구분용 이름표에 가까운 명목형

print(age + 5)  # 의미 있음: 5년 뒤 나이
print(room_number + "호")  # 의미 있음: 병실 표시
# print(room_number + 5)   # 의미 없음: 문자열과 숫자를 바로 더할 수 없음
```

### 잘못된 분석 예시

**범주형 변수를 숫자형으로 잘못 해석하는 경우**



```python
# 병실 번호 (명목형 - 숫자로 표현되지만 범주형)
room_numbers = [101, 205, 312, 108, 220, 101, 205, 101, 312, 207]
```

**❌ 잘못된 분석**


```python
# 평균 병실 번호를 계산하면?
average_room = sum(room_numbers) / len(room_numbers)
print(f"평균 병실: {average_room}호")
# → 187.2호라는 병실은 존재하지 않음! 의미 없는 계산
```

**✅ 올바른 분석**


```python
# ✅ 올바른 접근: 명목형으로 취급
import pandas as pd  # 데이터 분석 라이브러리

room_series = pd.Series(room_numbers)  # 리스트를 pandas Series로 변환
room_counts = room_series.value_counts()  # 각 값의 빈도수 계산 (내림차순 정렬)
most_used_room = room_series.mode()[0]  # 최빈값(가장 자주 나타나는 값) 추출

print("병실 사용 현황:")
print(room_counts)
print(f"가장 자주 사용되는 병실: {most_used_room}호")
# → 실제 존재하는 병실 번호로 의미 있는 분석
```

**환자 만족도 분석**



```python
# 만족도 점수 데이터 (순서형 - 숫자로 표현되지만 범주형)
satisfaction_scores = [
    1,
    2,
    3,
    4,
    5,
    3,
    4,
    4,
    5,
    3,
]  # 1:매우불만, 2:불만, 3:보통, 4:만족, 5: 매우만족

# 단순히 만족도 수준을 나타내는 라벨이지, 실제 수치적 의미를 갖는 숫자가 아님
```

```python
# ❌ 잘못된 접근: 연속형으로 취급
average_satisfaction = sum(satisfaction_scores) / len(satisfaction_scores)
print(f"평균 만족도: {average_satisfaction:.1f}점")
# → 3.4점이라는 선택지는 설문에 존재하지 않음!
```

```python
# ✅ 올바른 접근: 순서형으로 취급
satisfaction_series = pd.Series(satisfaction_scores)  # 리스트를 pandas Series로 변환
median_satisfaction = satisfaction_series.median()  # 중간값 계산 (순서형에서 중심 경향 측정)
mode_satisfaction = satisfaction_series.mode()[0]  # 최빈값 추출 (가장 자주 선택된 만족도)
satisfaction_dist = satisfaction_series.value_counts().sort_index()  # 빈도수를 점수 순서대로 정렬

print("만족도 분포:")
print(satisfaction_dist)
print(f"중간 만족도: {median_satisfaction}점")
print(f"가장 많은 응답: {mode_satisfaction}점")
# → 실제 선택 가능한 점수로 의미 있는 해석
```

## 데이터 타입에 따른 분석 방법

데이터 타입을 정확히 파악하면 적절한 분석 방법을 선택할 수 있습니다.



| 구분 | 숫자형 데이터 (Numerical) | 범주형 데이터 (Categorical) |
|------|---------------------------|---------------------------|
| **정의** | 숫자로 표현되며 수학적 계산이 가능한 데이터 | 그룹이나 범주로 분류되는 질적 데이터 |
| **의료 예시** | 나이(45세), 체중(70kg), 혈압(120mmHg), BMI(22.5) | 성별(남/여), 혈액형(A/B/O/AB), 진료과(내과/외과) |
| **특징** | • 수치적 의미<br>• 대소 비교 가능<br>• 연산 결과가 의미 있음 | • 분류/구분 목적<br>• 라벨링 역할<br>• 연산 결과가 무의미 |
| **가능한 분석** | 평균, 중간값, 표준편차, 상관관계 | 빈도, 비율, 최빈값, 교차분석 |
| **시각화** | 히스토그램, 산점도, 박스플롯 | 막대그래프, 파이차트 |
| **판단 기준** | "평균을 구하는 것이 의미가 있는가?" → **Yes** | "평균을 구하는 것이 의미가 있는가?" → **No** |
| **세부 분류** | • 연속형: 키 175.5cm<br>• 이산형: 환자 5명 | • 명목형: 혈액형 A형<br>• 순서형: 만족도 상/중/하 |





### 타입별 특징

- **연속형 데이터**: 모든 값이 수학적으로 의미가 있고, 값들 사이의 거리가 일정하므로 평균과 표준편차를 통해 분포의 중심과 퍼짐을 정확히 측정할 수 있습니다.

- **이산형 데이터**: 정수값만 존재하고 개수를 세는 특성상, 최빈값(가장 자주 나타나는 값)과 범위가 데이터의 특성을 더 잘 나타냅니다.

- **명목형 데이터**: 범주 간에 순서가 없어 평균이나 중간값이 의미가 없으므로, 각 범주의 빈도와 비율을 통해서만 분포를 파악할 수 있습니다.

- **순서형 데이터**: 순서는 있지만 범주 간 간격이 일정하지 않아 평균보다는 중간값이 더 안정적이고 의미 있는 중심 경향을 나타냅니다.



### 타입별시각화 방법

| 데이터 타입 | 적합한 그래프 | 부적합한 그래프 |
|------------|--------------|----------------|
| **연속형** | 히스토그램, 박스플롯, 산점도 | 막대그래프 |
| **이산형** | 막대그래프, 점 그래프 | 히스토그램 |
| **명목형** | 막대그래프, 파이차트 | 히스토그램, 선 그래프 |
| **순서형** | 막대그래프(순서대로), 누적그래프 | 파이차트 |
