# 데이터 시각화

## 데이터 시각화란 ?

데이터 시각화는 숫자로 이루어진 데이터를 그래프나 차트로 변환하여 패턴과 인사이트를 쉽게 발견할 수 있게 해주는 과정입니다.

**의료분야에서의 중요성**
- **환자 상태 모니터링**: 혈압, 혈당 변화 추이
- **질병 패턴 분석**: 연령대별, 성별 질병 발생률
- **치료 효과 평가**: 치료 전후 수치 비교
- **의료진 소통**: 복잡한 데이터를 직관적으로 전달




## Matplotlib vs Seaborn

### Matplotlib (기초 라이브러리)
**Matplotlib**은 파이썬에서 가장 기본이 되는 시각화 라이브러리로, 그래프의 모든 요소(색상, 크기, 축, 레이블 등)를 세밀하게 제어할 수 있는 저수준 인터페이스를 제공합니다.

- **특징**: 파이썬 시각화의 기초, 모든 것을 수동으로 설정
- **장점**: 완전한 커스터마이징 가능
- **단점**: 코드가 길고 복잡

### Seaborn (통계 시각화 전문)  
**Seaborn**은 Matplotlib을 기반으로 구축된 고수준 통계 시각화 라이브러리로, 복잡한 통계 차트를 간단한 코드 한 줄로 생성할 수 있게 해줍니다.   
의료 및 과학 분야에서 자주 사용하는 히스토그램, 박스플롯, 히트맵 등을 제공하며, 그룹별 비교나 상관관계 분석 같은 통계적 기능이 내장되어 있습니다.

- **특징**: Matplotlib 기반의 고급 통계 시각화 라이브러리
- **장점**: 간단한 코드, 아름다운 기본 스타일, 통계 기능 내장
- **단점**: 커스터마이징에 제한

**의료 데이터 분석에서는 Seaborn을 90% 이상 사용합니다!**

### 한글 폰트 / 라이브러리 첫 세팅

아래 셀은 **Windows, macOS, Linux, Google Colab**에서 공통으로 실행할 수 있도록
운영체제에 설치된 한국어 폰트를 자동으로 선택합니다.

- Windows: `Malgun Gothic` / `맑은 고딕` 우선 사용
- macOS: `AppleGothic` 우선 사용
- Linux·Colab: `NanumGothic`, `NanumBarunGothic`, `Noto Sans CJK KR` 계열 사용

````python
import platform
import warnings

import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

font_candidates = [
    "Malgun Gothic",  # Windows
    "맑은 고딕",  # Windows localized name
    "AppleGothic",  # macOS
    "NanumGothic",  # Linux / Colab
    "NanumBarunGothic",  # Linux / Colab
    "Noto Sans CJK KR",  # Linux
    "Noto Sans KR",  # Linux
]

available_fonts = {font.name for font in fm.fontManager.ttflist}
korean_font = next((font for font in font_candidates if font in available_fonts), None)

if korean_font:
    plt.rcParams["font.family"] = korean_font
    sns.set_theme(font=korean_font)
    print(f"한글 폰트 설정: {korean_font} ({platform.system()})")
else:
    print("한글 폰트를 자동으로 찾지 못했습니다.")
    print("Windows는 '맑은 고딕', macOS는 'AppleGothic' 설치 여부를 확인하세요.")
    print("Colab/Linux는 fonts-nanum 설치 후 런타임을 재시작하세요.")
    sns.set_theme()

plt.rcParams["axes.unicode_minus"] = False
warnings.filterwarnings("ignore")

print("라이브러리 로딩 완료!")
````

### 선 그래프 (Line Plot)

시간에 따른 변화를 표시할 때 많이 사용 됩니다.

````python
# 환자 혈압 변화 데이터
days = [1, 2, 3, 4, 5, 6, 7]  # 치료 일수
blood_pressure = [140, 138, 135, 132, 130, 128, 125]  # 수축기 혈압

# 기본 스타일 차트 그리기
plt.plot(days, blood_pressure)  # x축: 치료 일수 ,y축: 수축기 혈압

plt.show()  # 차트를 화면에 출력
````

**📋 Matplotlib 핵심 함수와 파라미터**

| 함수 | 역할 | 주요 파라미터 | 설명 | 예시 |
|------|------|---------------|------|------|
| `plt.figure()` | 차트 캔버스 생성 | `figsize` | 차트 크기 (가로, 세로) | `figsize=(10, 6)` |
| `plt.plot()` | 선 그래프 그리기 | `marker`, `color`, `linewidth` | 마커, 색상, 선 두께 | `marker='o', color='red'` |
| `plt.bar()` | 막대 그래프 그리기 | `color`, `alpha`, `width` | 색상, 투명도, 막대 너비 | `alpha=0.7, width=0.8` |
| `plt.title()` | 제목 설정 | `fontsize` | 제목 글자 크기 | `fontsize=14` |
| `plt.grid()` | 격자 표시 | `alpha`, `linestyle` | 격자 투명도, 선 스타일 | `alpha=0.3, linestyle='--'` |






````python
# 환자 혈압 변화 데이터
days = [1, 2, 3, 4, 5, 6, 7]  # 치료 일수
blood_pressure = [140, 138, 135, 132, 130, 128, 125]  # 혈압 수치
````

````python
# 차트 그리기
plt.figure(figsize=(8, 5))  # 새로운 차트 생성 및 크기 설정

# 선 그래프 그리기
plt.plot(
    days,
    blood_pressure,  # x, y 데이터
    marker="o",  # 점 모양 표시
    color="red",  # 선 색상
    linewidth=2,
)  # 선 두께

# 차트 꾸미기
plt.title("환자 혈압 변화 추이", fontsize=14)  # 제목 및 글자 크기
plt.xlabel("치료 일수", fontsize=12)  # x축 라벨
plt.ylabel("수축기 혈압 (mmHg)", fontsize=12)  # y축 라벨
plt.grid(True, alpha=0.5)  # 격자 표시, 투명도

plt.show()  # 차트를 화면에 출력
````

### 1. 빈도 그래프 (Countplot) - 범주형 데이터 분포

범주형 변수의 각 카테고리별 빈도수를 막대 그래프로 시각화하는 차트입니다.  
의료 AI에서는 환자군 구성비 파악, 질병 유병률 조사, 인구학적 특성 분석에 필수적으로 사용됩니다.

**의료 활용**: 환자 구성과 질병 분포 파악  
**임상 목적**: 병원 자원 배분, 환자군 특성 이해, 유병률 조사

**언제 사용할까?**
- 성별/연령대별 환자 분포 → 병원 진료과 배치 계획
- 질병별 발생 빈도 → 예방 프로그램 우선순위 결정
- 지역별 환자 분포 → 의료 서비스 접근성 개선



#### palette 색상 테마 종류

```python
# [Tip] palette 색상 테마 종류
# 1. 범주형 (구분): 'Set2', 'Paired', 'deep'       -> 서로 다른 그룹을 구분할 때 (예: 성별, 혈액형)
# 2. 연속형 (순서): 'viridis', 'Blues', 'Reds'     -> 크기나 강도가 커질 때 (예: 나이, BMI)
# 3. 발산형 (대조): 'RdYlBu', 'coolwarm', 'bwr'    -> 양극단으로 나뉠 때 (예: 정상 vs 당뇨, 긍정 vs 부정)
```

### 2. 히스토그램 (Histplot + hue) - 그룹별 분포 비교 ⭐⭐⭐⭐

변수의 분포를 막대로 표현하고, hue 파라미터를 통해 **그룹별로 분포를 비교**할 수 있는 차트입니다.     
의료 AI에서는 환자군별 특성 차이 분석, 위험군 식별, 치료 효과 모니터링에 광범위하게 활용됩니다.

**의료 활용**: 그룹 간 분포 차이 분석 및 패턴 발견  
**임상 목적**: 환자군 특성 비교, 위험군 식별, 치료 효과 평가

**언제 사용할까?**
- **연속형 데이터**: 당뇨병군 vs 정상군의 BMI 분포 비교 → 위험 BMI 구간 파악
- **순위형 데이터**: 치료군별 건강상태 점수 분포 → 치료 효과 시각적 확인  
- **그룹별 패턴**: 연령대별 혈압 분포 → 연령 특화 치료 기준 수립

**핵심**: 두 개 이상의 그룹을 **나란히 비교**하여 분포의 차이와 겹치는 구간을 한눈에 파악할 수 있습니다.


````python
# BMI 분포 히스토그램 (당뇨병 여부별)
plt.figure(figsize=(12, 5))  # 2개 차트용 캔버스

#  범례 표기를 위해 0/1 숫자를 한글로 변환
df["Label"] = df["Diabetes_binary"].replace({0.0: "정상", 1.0: "당뇨병"})  # 여러개를 한꺼번에 변환

# 1. 첫 번째 차트: BMI 분포 분석
plt.subplot(1, 2, 1)  # 1행 2열 중 첫 번째 칸 선택
sns.histplot(
    data=df,
    x="BMI",  # 분석할 데이터
    hue="Label",  # 비교 그룹 (정상 vs 당뇨)
    bins=30,  # 막대 개수 (데이터 세분화 정도)
    alpha=0.7,  # 투명도
    palette="Reds",
)  # 색상 테마

plt.title("당뇨병 여부별 BMI 분포")
plt.xlabel("BMI (체질량지수)")
plt.ylabel("환자 수")

# 2. 두 번째 차트: 건강상태 분포 분석
plt.subplot(1, 2, 2)  # 1행 2열 중 두 번째 칸 선택
sns.histplot(
    data=df,
    x="GenHlth",  # 분석할 데이터 (순위형 점수)
    hue="Label",  # 비교 그룹
    bins=5,  # 점수 범위 (1~5)
    alpha=0.7,  # 투명도
    palette="viridis",  # 색상 테마
    discrete=True,
)  # 정수값 처리 (막대를 숫자에 딱 맞춤)

plt.title("당뇨병 여부별 전반적 건강상태")
plt.xlabel("건강상태 (1=매우좋음 ~ 5=매우나쁨)")
plt.ylabel("환자 수")

# 차트 위치 조정
plt.tight_layout()  # 레이아웃 최적화
plt.show()  # 차트를 화면에 출력

# ==============================  통계 분석으로 차이 확인  ==============================
# 1. BMI 분포 분석
print("\n📊 BMI 분포 분석:")

# 비만(BMI≥30) 비율
# apply: pandas에서 반복문을 사용하지 않고 모든 데이터에 원하는 함수 적용
# (x: 그룹핑된 각각의 Series, 조건마스크: 조건에 부합하는 bool마스크의 True는 1로 변환되어 연산)
obesity_rate = df.groupby("Diabetes_binary")["BMI"].apply(
    lambda x: (x >= 30).mean() * 100
)  # 비만율
print(f"정상인 중 비만 비율: {obesity_rate[0]:.1f}%")
print(f"당뇨병 환자 중 비만 비율: {obesity_rate[1]:.1f}%")

# 2. 건강상태 분포 분석
print("\n📊 건강상태(1:우수 ~ 5:나쁨) 분포 분석:")

# GenHlth가 4점 이상인 비율 계산
poor_health_rate = df.groupby("Diabetes_binary")["GenHlth"].apply(lambda x: (x >= 4).mean() * 100)
print(f"정상인 중 건강 나쁨 응답: {poor_health_rate[0]:.1f}%")
print(f"당뇨 환자 중 건강 나쁨 응답: {poor_health_rate[1]:.1f}%")
````

### 3. 박스플롯 (Boxplot) - 분포와 이상치 확인 ⭐⭐⭐⭐⭐

데이터의 중심 경향(중앙값), 분산(사분위수 범위), 이상치를 한 번에 보여주는 통계 차트입니다. 의료 AI에서는 생체 지표의 정상 범위 설정, 극단적 환자 케이스 식별, 치료군별 효과 비교에서 가장 신뢰받는 시각화 방법입니다.

**의료 활용**: 수치 분포의 중심 경향과 이상치 탐지  
**임상 목적**: 정상 범위 설정, 극단적 케이스 식별, 그룹 간 비교

**언제 사용할까?**
- 검사 수치의 정상 범위 → 진단 기준 설정, 참고치 구간 정의
- 치료군별 효과 비교 → 약물 효과성 평가, 최적 치료법 선택
- 이상치 환자 식별 → 특별 관리 대상, 추가 검사 필요 환자



**박스플롯 구성 요소 가이드**

1. 박스 중앙의 선: 중앙값 (Median)
2. 박스 범위: 제3사분위수(Q3) / 제1사분위수(Q1) -> 데이터의 50%가 모여있는 구간
3. 수염 (Whiskers): 정상 범위의 최소/최대값
4. 흩뿌려진 점들: 이상치 (Outliers) -> 일반적인 범위를 벗어난 특별한 사례

````python
# 차트 크기 및 스타일 설정
plt.figure(figsize=(12, 5))

# 1. 첫 번째 차트: 당뇨병 여부별 BMI 분포
plt.subplot(1, 2, 1)
sns.boxplot(
    data=df,
    x="Diabetes_binary",  # x축: 그룹 구분
    y="BMI",  # y축: BMI 데이터
    hue="Diabetes_binary",  # 색상 구분
    palette="coolwarm",  # 색상 테마
    legend=False,
)  # 범례 숨김
plt.title("당뇨병 여부별 BMI 분포")
plt.xlabel("진단 여부")
plt.ylabel("BMI (체질량지수)")
plt.xticks([0, 1], ["정상", "당뇨병"])
# 💡해석: 정상군에 비해 당뇨군의 BMI 분포가 오른쪽(높은 수치)으로
# 치우쳐 있어, 비만이 주요 위험 인자임을 확인 가능

# 2. 두번째 차트: 당뇨병 여부별 연령대 분포
# 연령대 라벨 리스트 생성 (1~13 매핑)
age_labels = [
    "18-24",
    "25-29",
    "30-34",
    "35-39",
    "40-44",
    "45-49",
    "50-54",
    "55-59",
    "60-64",
    "65-69",
    "70-74",
    "75-79",
    "80+",
]

plt.subplot(1, 2, 2)
sns.boxplot(
    data=df,
    x="Diabetes_binary",  # x축: 그룹 구분
    y="Age",  # y축: 연령대 (1~13단계)
    hue="Diabetes_binary",  # 색상 구분
    palette="pastel",  # 색상 테마
    legend=False,
)  # 범례 숨김

plt.title("당뇨병 여부별 연령대 분포")
plt.xlabel("진단 여부")
plt.ylabel("연령대")
plt.xticks([0, 1], ["정상", "당뇨병"])
plt.yticks(ticks=range(13), labels=age_labels, rotation=45)
# 💡해석: 당뇨군은 정상군보다 박스가 전체적으로 상단에 위치하여,
# 연령대가 높을수록 당뇨 발병 비중이 커지는 '연령의 영향력' 확인 가능

plt.tight_layout()
plt.show()

# ==============================  통계 분석으로 차이 확인  ==============================


# 1. BMI 중앙값 차이 (차트 1)
print("\n📊 당뇨병 여부별 BMI 중앙값 분석:")
print(df.groupby("Diabetes_binary")["BMI"].median())

# 2. 고령층의 당뇨병 발병률 (차트 2)
print("\n📊 고령층 당뇨병 유병률 분석:")

senior_df = df[df["Age"] >= 9]  # 65세 이상 그룹
senior_diabetes_rate = senior_df["Diabetes_binary"].mean() * 100

print(f"고령층(65세 이상) 내 당뇨병 환자 비율: {senior_diabetes_rate:.1f}%")
````

**해석 가이드**

박스플롯의 비대칭은 **'데이터가 어디로 쏠려 있는가'를** 보여주는 이정표입니다.

- 위쪽 수염이 길다: 높은 값들이 넓게 퍼져 있다.

- 아래쪽 수염이 길다: 낮은 값들이 넓게 퍼져 있다.

- 박스 안의 선(중앙값)도 한쪽으로 쏠려 있다: 중간 50%의 데이터조차 특정 방향으로 치우쳐 있다.

````python
# 당뇨병 여부별 신체 건강 불량일수
plt.figure(figsize=(8, 5))
sns.boxplot(
    data=df,
    x="Diabetes_binary",  # x축: 당뇨병 진단 여부
    y="PhysHlth",  # y축: 최근 30일간 몸이 아팠던 일수 (0~30일)
    hue="Diabetes_binary",  # 색상 구분
    palette="magma",  # 색상 테마
    legend=False,
)  # 범례 숨김
plt.title("당뇨병 여부별 신체 건강 불량일수")
plt.xlabel("진단 여부")
plt.ylabel("아픈 날짜 (최근 30일)")
plt.xticks([0, 1], ["정상", "당뇨병"])
# 💡해석: 대부분 0(정상)이라 박스가 납작하지만, 당뇨군에서
# 3사분위수(박스 높이)가 더 높게 나타나 신체적 불편을 겪는 빈도가
# 더 잦음을 확인 가능
# 상단의 점들은 만성 질환자들의 특이 케이스

# 데이터의 50% 이상이 0일 때 박스플롯의 박스는 바닥(0)에 붙게 됩니다.

plt.show()

# ==============================  통계 분석으로 차이 확인  ==============================

# 3. 신체 건강 불량일수 0일인 사람의 비율 (차트 3)
print("\n📊 신체 건강상태 분포 분석:")
perfect_health_rate = (df["PhysHlth"] == 0).mean() * 100
print(f"전체 응답자의 {perfect_health_rate:.1f}%가 0일 아픔")
````

### 4. 히트맵 (Heatmap) - 상관관계 분석 ⭐⭐⭐⭐

변수들 간의 상관계수를 색상으로 표현하여 데이터 내 변수들의 연관성을 한눈에 파악할 수 있는 차트입니다.   
의료 AI에서는 질병 위험 요인 간의 복잡한 연관성을 분석하고, 예측 모델에 사용할 핵심 변수를 선별하는데 필수적인 도구입니다.

**의료 활용**: 질병 위험 요인들 간의 연관성 분석  
**임상 목적**: 복합 질환 관리, 위험 요인 우선순위 결정, 예방 전략 수립

**언제 사용할까?**
- 생활습관과 질병의 연관성 → 예방 교육 프로그램 개발
- 검사 수치들 간의 관계 → 핵심 모니터링 지표 선정
- 약물 부작용 패턴 → 안전한 약물 조합 가이드라인

**핵심**: 수치형 변수들 간의 선형 상관관계 강도와 방향을 색상 농도로 표현하여, 핵심 변수 선별에 활용됩니다.

````python
# 당뇨병 데이터의 주요 의료 변수들 간 상관관계 분석
medical_vars = [
    "Diabetes_binary",  # 타겟 변수
    "BMI",
    "HighBP",  # 생리학적 지표
    "Stroke",
    "HeartDiseaseorAttack",  # 질병력
    "PhysActivity",
    "Smoker",  # 생활습관
    "Sex",
    "Age",  # 인구학적
]

correlation_matrix = df[medical_vars].corr()  # 피어슨 상관계수 계산

correlation_matrix
````

````python
# 히트맵 차트

plt.figure(figsize=(10, 6))  # 히트맵용  캔버스
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))  # 상삼각 마스킹

sns.heatmap(
    correlation_matrix,  # 상관계수 데이터프레임
    mask=mask,  # 중복 제거용 마스크
    annot=True,  # 숫자 표시
    cmap="RdBu_r",  # 빨강-파랑 색상맵
    center=0,  # 0 중심 색상
    square=True,  # 정사각형 셀
    fmt=".2f",  # 소수점 2자리
    cbar_kws={"label": "상관계수", "shrink": 0.8},
)  # 컬러바 설정

plt.title("당뇨병 데이터: 의료 변수들 간의 상관관계", fontsize=18, pad=20)
plt.xticks(rotation=45, ha="right", fontsize=10)  # x축 라벨 회전
plt.yticks(rotation=0, fontsize=10)  # y축 라벨
plt.tight_layout()  # 레이아웃 최적화
plt.show()  # 차트를 화면에 출력


# ==============================  통계 분석으로 차이 확인  ==============================

# 당뇨병과 상관관계 높은 변수 분석
diabetes_corr = correlation_matrix["Diabetes_binary"].abs().sort_values(ascending=False)
print("\n📊 당뇨병과 상관관계가 높은 변수 TOP 5:")
print("=" * 50)
for i, (var, corr) in enumerate(diabetes_corr[1:6].items(), 1):  # 자기 자신 제외
    print(f"{i:2d}. {var:<20}: {corr:.3f}")
````
